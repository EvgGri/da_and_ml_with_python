# Задание 1

В качестве исследуемого распределения было выбрано распределение хи квадрат (ввиду моих текущих задач). Сгенерируем из него выборку объёма 1000, далее построим гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения случайной величины.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import chi2
%matplotlib inline
```

Распределе́ние $\chi ^2$ (хи-квадра́т) с $k$ степеня́ми свобо́ды — это распределение суммы квадратов $k$ независимых стандартных нормальных случайных величин.

```python
df = 55 # число степеней свободы, равное математическому ожиданию распределения
```

```python
r = chi2.rvs(df, size=1000) # формирование необходимой выборки
```

```python
x = np.linspace(20, 90, 1000) # определение диапазона для Х
```

```python
plt.plot(x, chi2.pdf(x, df), 'r-', lw=5, alpha=0.6, label='Theoretical pdf')
plt.hist(r, normed=True, histtype='stepfilled', alpha=0.2, label='Experimental pdf')
plt.legend(loc='best', frameon=False)
plt.title("Propability density function for $\chi^2$ ")
```

# Задание 2

Оценим распределение выборочного среднего случайной величины при разных объёмах выборок.

В начале построим график нормального распределения случайной величины с с с $\mu=55$ и $\sigma^2=55$ (у рассматриваемого распределения $\mu$ = $\sigma^2$)

```python
xx = np.linspace(20, 90, 10000) # задание диапазона для графика
```

```python
norm_rv = sts.norm(df, np.sqrt(2 * df)) # формирование нормального распределения с указанными параметрами
```

```python
plt.plot(xx, norm_rv.pdf(xx), 'r-', lw=5, alpha=0.6, label='Theoretical pdf')
plt.title("Propability density function for normal distribution ")
```

## Размер выборки равен 10

```python
n = 10
```

Нижняя граница 95% доверительного интервала для среднего значения, при этом объеме выборки:

```python
calculatedMeanLowBound = df - 2 * np.sqrt(df)/np.sqrt(n) 
calculatedMeanLowBound
```

Верхняя граница 95% доверительного интервала для среднего значения, при этом объеме выборки:

```python
calculatedMeanHighBound = df + 2 * np.sqrt(df)/np.sqrt(n) 
calculatedMeanHighBound
```

Ширина 95% доверительного интервала:

```python
calculatedMeanHighBound - calculatedMeanLowBound
```

Построим график распределения выборочного среднего случайной величины при $n=10$

```python
sampleMean = list()
for i in range(1, 1000):
    rI = chi2.rvs(df, size=n)
    sampleMean.append(np.mean(rI))
sampleMeanScatterPlot = pd.DataFrame(sampleMean, columns=['Experimental plot'])
ax = sampleMeanScatterPlot.plot(kind='density')
plt.hist(sampleMean, normed=True, histtype='stepfilled', alpha=0.2, label='Experimental pdf')
plt.legend(loc='best', frameon=False)
plt.title("Propability density function for for $\chi^2$ with $n=10$")
```
